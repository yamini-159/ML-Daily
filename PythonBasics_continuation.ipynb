List Comprehension

```python
numbers = [1,2,3,4,5]
squares=[]

for x in numbers:
    squares.append(x**2)
print(squares)
```

```python
squares = [x**2 for x in range (5)]
print(squares)
```

```python
even = [x for x in range (20) if x%2==0]
print(even)
```

```python
words = ["python","machinelearning","datascience"]
upper = [w.upper() for w in words]
print(upper)
```
