```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
```

```python
fig, ax=plt.subplots()
ax.plot(x,y, label="sin(x)",color="green")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.set_title("Sine_Wave")
ax.legend()
plt.show()
```

```python
np.random.seed(42)
```

```python
epochs=np.arange(1,21)
train_loss = np.exp(-epochs/5)+np.random.rand(20)*0.5
```

```python
val_loss=np.exp(-epochs/6)+np.random.rand(20)*0.08
```

```python
plt.plot(epochs,train_loss,label="train")
plt.plot(epochs,val_loss,label="validity")
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.title("Training vs Validation loss")
plt.legend()
plt.show()
```

```python
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
```

```python
iris.head()
```

```python
sns.histplot(data=iris,x="sepal_length",hue="species",kde=True)
```

```python
sns.boxplot(data=iris,x="species",y="petal_length")
plt.title("Box=plot")
plt.show()
```

```python
sns.violinplot(data=iris,x="species",y="petal_width")
plt.title("Violin plot - distribution shape")
plt.show()
```

```python
corr = iris.drop(columns=["species"]).corr()
sns.heatmap(corr,annot=True)
```
